```python
import pandas as pd
import csv 
import glob
```

```python
# Define the path to the data folder
data_folder = 'data/epa-airtoxscreen/'

# Define a list of years
years = [2014, 2017, 2018, 2019]

# Initialize empty lists to store DataFrames
ambient_dfs = []
cancer_dfs = []
exposure_dfs = []

# Loop through each year
for year in years:
    # Read CSV files into DataFrames
    ambient_df = pd.read_csv(f'{data_folder}{year}_Ambient_Concentration_ug_m3.csv').head(10)
    cancer_df = pd.read_csv(f'{data_folder}{year}_Cancer_Risk_in_a_million_and_Noncancer_Risk_hazard_quotient.csv').head(10)
    exposure_df = pd.read_csv(f'{data_folder}{year}_Exposure_Concentration_ug_m3.csv').head(10)
    
    # Add 'Year' column to each DataFrame
    ambient_df['Year'] = year
    cancer_df['Year'] = year
    exposure_df['Year'] = year
    
    # Append DataFrames to the lists
    ambient_dfs.append(ambient_df)
    cancer_dfs.append(cancer_df)
    exposure_dfs.append(exposure_df)

# Initialize empty list to store merged DataFrames for each year
merged_dfs = []

# Perform merge separately for each year
for ambient_df, cancer_df, exposure_df in zip(ambient_dfs, cancer_dfs, exposure_dfs):
    merged_df = pd.merge(pd.merge(ambient_df, cancer_df, on='FIPS', how='outer'), exposure_df, on='FIPS', how='outer')
    merged_dfs.append(merged_df)

# Concatenate all DataFrames in merged_dfs list
epa_merged = pd.concat(merged_dfs, ignore_index=True)

display(epa_merged.head())
```
